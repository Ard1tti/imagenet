```python
import pickle

f = open('../../data/ILSVRC2012_data.txt', 'r')
data = pickle.load(f)
f.close()
```

```python
nwid_list=list(set([d[1] for d in data]))
nwid_dict={x:i for i,x in enumerate(nwid_list)}
len(nwid_dict)
```

```python
files=np.asarray([val for val in os.listdir("../../data/images")])
labels=np.asarray([nwid_dict[d[1]] for d in data])
```

```python
train_data=Input(files,labels,capacity=50,class_num=1000,IMG_DIR='../../data/images/')
image, label = train_data.dequeue()
```

```python
with tf.Session() as sess:
    threads = tf.train.start_queue_runners(sess=sess)
    for i in xrange(3):
        res_image, res_lab = sess.run([image, label])
        print(res_image.shape, res_lab)
```
