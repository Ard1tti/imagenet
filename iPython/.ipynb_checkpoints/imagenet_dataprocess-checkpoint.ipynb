```python
f = open("wnid_data.txt","r")
temp = f.read()
temp = temp.split(" &nbsp\n")[:-1]
class_meta = {w[22:31]: w[33:-4] for w in temp}
class_list = [w[22:31] for w in temp]
```

```python
f = open("urls/imagenet_urls.txt","r")
lines = f.readlines()
```

```python
whole = [(l.split("\t")[1][:-1],l.split("\t")[0][:9]) for l in lines]
```

```python
data = [w for w in whole if w[1] in class_list]
```

```python
import pickle
f = open('urls/ILSVRC2010_urls.txt', 'w')
pickle.dump([data, class_list, class_meta], f)
f.close()
```
