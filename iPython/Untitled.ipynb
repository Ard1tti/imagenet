```python
val_list=[val.split(".")[0] for val in os.listdir("images/")]
val_list
```

```python
import xml.etree.ElementTree as ET

nwid = []
for f in val_list:
    doc = ET.parse("imagenet-images/val/"+f+".xml")
    root=doc.getroot()
    nwid.append(root.find("object").find("name").text)
```
