```python
from imagenet_input_ed import Input
import numpy as np
import pickle

with open('../../data/urls/ILSVRC2010_data.txt') as f:
    files, labels, class_list, _ = pickle.load(f)
```

```python
class_files=[[files[i] for i in xrange(len(labels))
              if labels[i] is j] for j in xrange(1000)]
```

```python
val_files=[]
val_labels=[]
train_files=[]
train_labels=[]
index=[0]*1000
for f, l in zip(files,labels):
    if index[l] < 50:
        val_files.append(f)
        val_labels.append(l)
        index[l]=index[l]+1
    else:
        train_files.append(f)
        train_labels.append(l)
        
```

```python
with open('../../data/urls/ILSVRC2010_train.txt','w') as f:
    pickle.dump([train_files, train_labels, class_list], f)
    
with open('../../data/urls/ILSVRC2010_val.txt','w') as f:
    pickle.dump([val_files, val_labels, class_list], f)
```
