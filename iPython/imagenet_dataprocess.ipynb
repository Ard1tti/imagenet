```python
import Image
import numpy as np
from StringIO import StringIO
def PIL2array(img):
    return np.array(img.getdata(), np.uint8).reshape([img.size[0],img.size[1],-1])
```

```python
import pickle
import urllib

f = open('urls/ILSVRC2010_urls.txt', 'r')
data, class_list, class_meta = pickle.load(f)
f.close()

url=[d[0] for d in data]
```

```python
for u in url[0:10]:
    f=urllib.urlopen(u)
    if f.info().type in ['image/jpg','image/jpeg','image/pjpeg',
                        'image/gif','image/png']:
        img=Image.open(StringIO(f.read()))
        image=PIL2array(img)
        print(image.shape)
```
